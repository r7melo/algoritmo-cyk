# Algoritmo CYK (Cocke-Younger-Kasami)


## 1. Descrição do Problema
O objetivo deste trabalho é implementar o **Algoritmo CYK** (Cocke-Younger-Kasami), um método de análise sintática de ordem polinomial $O(n^3)$.

Ele deve ser capaz de reconhecer se uma palavra $w$ é gerada por uma **Gramática Livre de Contexto (GLC)** que esteja na **Forma Normal de Chomsky (FNC)**.

**Entrada:**
- Um JSON com a estrutura da gramática válida.
- Uma palavra $w$.
- 
**Saída:**
- Uma mensagem de validação se a palavra é pertencente a linguagem gerada pela gramática da fornecida.

## 2. O Algoritmo
O CYK utiliza **Programação Dinâmica** para preencher uma tabela (ou matriz triangular) de baixo para cima.

### Estrutura de Dados
Uma matriz tridimensional (ou bidimensional de conjuntos) $T[n, n, |V|]$, onde:
- **Linhas ($i$):** Representam o *tamanho* da subpalavra (de 1 a $n$).
- **Colunas ($j$):** Representam a *posição de início* da subpalavra.
- **Célula:** Contém as variáveis que geram aquela subpalavra.

### Passo a Passo
1.  **Base (Comprimento 1):**
    Para cada letra da palavra, verificamos quais regras $V \to \text{terminal}$ geram aquela letra.
2.  **Passo Indutivo (Comprimento 2 até $n$):**
    Para tamanhos maiores, dividimos a subpalavra em duas partes em todos os pontos de corte possíveis ($k$).
    Verificamos se existe uma regra $A \to BC$ tal que $B$ gera a primeira parte e $C$ gera a segunda.
3.  **Verificação Final:**
    Se a célula do topo da pirâmide ($T[1, n]$) contiver o símbolo de partida $S$, a resposta é **'$ w \in L(G)$'**.
